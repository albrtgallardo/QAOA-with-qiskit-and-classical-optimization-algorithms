```python
import numpy as np
import autograd as ag
import matplotlib.pyplot as plt
```

```python
T4 = 0.5
T3 = 0
T2 = -1
T1 = 0
T0 = 0

x = np.arange(-2, 2, 0.01)
y = T4*(x**4) + T3*(x**3) +T2*(x**2) + T1*x + T0

plt.figure()
plt.plot(x, y)
plt.grid()
```

```python
def f(x):
    y = T4*(x**4) + T3*(x**3) +T2*(x**2) + T1*x + T0
    return y
```

```python
def ini_vec(inf_range,sup_range):
    return round(np.random.uniform(inf_range,sup_range), 1)
```

```python
def gradient(v):
    grad_f = ag.grad(f)
    return (grad_f(np.array(v)))
```

```python
def gradient_descend(inf_range, sup_range, n_iter, learning_rate, tolerance):
    v = ini_vec(inf_range,sup_range)
    print(f"vector inicial {v}")
    for i in range(n_iter):
        g = gradient(v)
        print(f'El gradiente es {g}')
        v_new = v - learning_rate*g
        #print(v_new)

        if abs(v_new - v) < tolerance:
            break

        v = v_new
        
    return v
```

```python
print(f"Solucion {gradient_descend(-2, 2, 100, 0.01, 0.00001)}")
```
